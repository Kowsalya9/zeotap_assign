```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the datasets
customer = pd.read_csv("Customers.csv")
transaction = pd.read_csv("Transactions.csv")
product = pd.read_csv("Products.csv")

# Check for missing values
print(customer.isnull().sum())
print(product.isnull().sum())
print(transaction.isnull().sum())

#plot the univariant analysis 
#customer analysis
sns.countplot(x='Region', data= customer)
plt.show()
#product analysis
sns.countplot(x='Category', data=product)
plt.show()
sns.histplot(x='Price', data=product)
plt.show()
#transaction analysis
sns.countplot(x='Quantity', data= transaction)
plt.show()
```

```python
#plot the biviarant analysis
cust_prod = pd.merge(customer, transaction, on='CustomerID')
cust_trans = pd.merge(cust_prod, product, on = 'ProductID')

sns.boxplot(x='Region', y= 'TotalValue', data=cust_trans)
plt.show()
```
